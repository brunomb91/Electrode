
## O método, retirado de "Rhythmic activities of the brain: Quantifying the high complexity of beta and gamma oscillations during visuomotor tasks"; consiste em:

> - Calcular a CAR (Common Average Reference), subtração do sinal do eletrodo pelo sinal sinal que possui amplitude comum a todos os eletrodos

> - Filtrar o sinal utilizando a janela de Kaiser, considerando as ondas Alpha, Beta, Gamma, Delta e Theta. As bandas de EEG mostram a função cerebral e indicam a presença ou ausência de atividade cerebral específica em áreas específicas do cérebro.

> - Utilizar o plano entropia-complexidade para identificar atividades cerebrais e informações que são carregadas por diferentes oscilações rítmicas.

### Cálculo da janela de Kaiser utilizando o Scipy

```python
from scipy.signal import kaiserord, firwin, freqz
import matplotlib.pyplot as plt

fs = 110.0
cutoff = 50
width = 24

stopband = 60
transition = 0.1
passband = 0.05

numtaps, beta = kaiserord(60, width/(0.5*fs))
taps = firwin(numtaps, cutoff, window=('kaiser', beta),
              scale=True, nyq=0.5*fs)

w, h = freqz(taps, worN=8000)
w *= 0.5*fs/np.pi  # Convert w to Hz.

ideal = w < cutoff  # The "ideal" frequency response.
deviation = np.abs(np.abs(h) - ideal)
deviation[(w > cutoff - 0.5*width) & (w < cutoff + 0.5*width)] = np.nan

plt.plot(w, 20*np.log10(np.abs(deviation)))
plt.xlim(0, 0.5*fs)
plt.ylim(-90, -60)
plt.grid(alpha=0.25)
plt.axhline(-65, color='r', ls='--', alpha=0.3)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Deviation from ideal (dB)')
plt.title('Lowpass Filter Frequency Response')
plt.show()
```

### Cálculo da entropia-complexidade da média dos 64 eletrodos para os 109 indivíduos:

```python
import matplotlib.pyplot as plt
import numpy as np
from mne.io import read_raw_edf
from ordpy import complexity_entropy, maximum_complexity_entropy, minimum_complexity_entropy
import pandas as pd

data = read_raw_edf("Datasets/S001/S001R01.edf", preload=True)
raw = data.get_data()
channels = data.ch_names

d = {}

average = []

d['entropia'] = []
d['complexidade'] = []
d['eletrodo'] = []
```

```python
elec = 33

print("Importando dados dos eletrodos %d..." %(elec + 1))
elec_avg = np.array([read_raw_edf(f, preload=True).get_data()[elec] for f in paths()])

print("Calculando média...")
for j in range(elec_avg[0].shape[0]):
    s = 0
    for i in range(elec_avg.shape[0]):
        #print("i: %d\nj: %d" %(i, j))
        try:
            s = s + elec_avg[i][j]
        except:
            continue
    average.append(np.sqrt((s**2)/elec_avg[0].shape[0]))
    #average.append(s**2/elec_avg[0].shape[0])
        
h, c = complexity_entropy(average, dx=6)

d['eletrodo'].append(channels[elec])
d['entropia'].append(h)
d['complexidade'].append(c)

del(elec_avg)
print("Importando para CSV...")
df = pd.DataFrame(d)
df.to_csv('hc.csv', sep=';')

print(df)
```
